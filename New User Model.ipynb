```python
us_restaurant_review_lasvegas_nv.is_copy = False
us_restaurant_review_lasvegas_nv['category']=pd.Series()
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('Mexican'),'category']='Mexican'
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('French'),'category']='French'
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('Buffets'),'category']='Buffets'
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('Middle Eastern'),'category']='Middle Eastern'
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('Steakhouses'),'category']='American'
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('Pizza'),'category']='American'
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('Italian'),'category']='Italian'
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('Seafood'),'category']='Seafood'
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('Cantonese'),'category']='Chinese'
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('Chinese'),'category']='Chinese'
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('Thai'),'category']='Thai'
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('Noodles'),'category']='Noodles'
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('Fast Food'),'category']='Fast Food'
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('Southern'),'category']='American'
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('American'),'category']='American'
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('Japanese'),'category']='Japanese'
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('Korean'),'category']='Korean'
us_restaurant_review_lasvegas_nv.loc[us_restaurant_review_lasvegas_nv.categories.str.contains('Vietnamese'),'category']='Vietnamese'

us_restaurant_review_lasvegas_nv.category[:30]
```

```python
us_restaurant_review_lasvegas_nv = us_restaurant_review_lasvegas_nv.dropna(axis=0, subset=['category'])
us_restaurant_review_lasvegas_nv
```

```python
us_restaurant_review_lasvegas_nv.shape
```

```python
restaurant_features = pd.DataFrame( columns=['business_id', 'Japanese','Mexican','Italian','American','Chinese','Thai', 'Creperies','French'])
count = 0
while count < 2803:
    values = {'business_id':us_restaurant_review_lasvegas_nv.business_id.iloc[count],'Japanese':False,'Mexican':False,'Italian':False,'American':False,'Chinese':False,'Thai':False,'Creperies':False,'French':False}

    if count == 2803:
        break
   
    if us_restaurant_review_lasvegas_nv.categories.str.contains('Creperies').iloc[count] == True:
        values['Creperies'] = True
    if us_restaurant_review_lasvegas_nv.categories.str.contains('Middle Eastern').iloc[count] == True:
        values['French'] = True

    if us_restaurant_review_lasvegas_nv.categories.str.contains('French').iloc[count] == True:
        values['French'] = True
    if us_restaurant_review_lasvegas_nv.categories.str.contains('Japanese').iloc[count] == True: 
        values['Japanese'] = True
    if us_restaurant_review_lasvegas_nv.categories.str.contains('Mexican').iloc[count] == True: 
        values['Mexican'] = True
    if us_restaurant_review_lasvegas_nv.categories.str.contains('Italian').iloc[count] == True: 
        values['Italian'] = True
    if us_restaurant_review_lasvegas_nv.categories.str.contains('American').iloc[count] == True or  us_restaurant_review_lasvegas_nv.categories.str.contains('Pizza').iloc[count] == True: 
        values['American'] = True
    if us_restaurant_review_lasvegas_nv.categories.str.contains('Chinese').iloc[count] == True:
        values['Chinese'] = True
    if us_restaurant_review_lasvegas_nv.categories.str.contains('Thai').iloc[count] == True: 
        values['Thai'] = True
    if us_restaurant_review_lasvegas_nv.categories.str.contains('Thai').iloc[count] == True: 
        values['Thai'] = True

    restaurant_features =restaurant_features.append(values, ignore_index=True)
    
    count = count + 1    
```

```python
import pickle
from sklearn.externals import joblib
def new_user_model():
    restaurant_features = pd.read_csv("yelp_dataset/restaurant_features.csv")
    new_user_restaurant_jac_sim = pd.DataFrame( columns=['jac_sim','business_id'])
    
    for index, row in restaurant_features.iterrows():
        jac_sim = jaccard_similarity_score([False, True, False, False, False, False, False, False],[row['Japanese'], row['Mexican'],row['Italian'], row['American'],row['Chinese'], row['Thai'],row['Creperies'], row['French']])
        new_user_restaurant_jac_sim = new_user_restaurant_jac_sim.append({'jac_sim':jac_sim,'business_id':row['business_id']}, ignore_index=True)
    new_user_restaurant_jac_sim_asc = new_user_restaurant_jac_sim.sort_values(by=['jac_sim'], ascending=False)
    new_user_restaurant_jac_sim_asc_top30 = new_user_restaurant_jac_sim_asc[:30]
    return new_user_restaurant_jac_sim_asc_top30['business_id'].tolist()
#new_user_model()
```
